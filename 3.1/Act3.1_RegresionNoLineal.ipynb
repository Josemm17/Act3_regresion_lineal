## Carga de archivos y limpieza de datos

```python
#Cargamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score

df=pd.read_csv("Rio_limpio.csv")
```

```python
vars_num=df.select_dtypes(include=['float64', 'int64','float','int'])

Corr_factors=vars_num.corr()

#Obtener el valor absoluto
Corr_factors_abosluto=abs(Corr_factors)

#Ajustar el heatmap con los coeficientes de correlacion
Heat_map=sns.heatmap(Corr_factors_abosluto, cmap="coolwarm")
Heat_map
```

## Elegir 2 modelos de correlaci√≥n No lineal y Analizar la correlaci√≥n que existe entre las siguientes variables: 

->host_response_rate

->host_acceptance_rate

->host_total_listings_count

->accommodates

->reviews_per_month

->price


### host_response_rate

```python
#Declaramos las variables dependientes e independientes para la regresi√≥n No lineal
Vars_Indep= df[['availability_30']]
Var_Dep= df['host_response_rate']

#Redefinimos las variables 
x= Vars_Indep
y= Var_Dep

def  func1 (x, a, b, c):
    return a*x**2 + b*x + c #("Funci√≥n cuadr√°tica")

#Ajustamos los par√°metros de la funci√≥n curve_fit
parametros1, covs = curve_fit(func1, df['availability_30'], df['host_response_rate'])

# Predicciones
y_pred1 = func1(x, *parametros1)


#Graficamos las predicciones y los datos originales para realizar la comparaci√≥n
plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, y_pred1, label="y-predecida con y=a*x**2 + b*x + c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 

#Calculamos el coeficiente de determinaci√≥n del modelo
R2_Modelo1 = r2_score(y, y_pred1) 
print(R2_Modelo1)

#Calculamos el coeficiente de correlaci√≥n del modelo
R = np.sqrt(abs(R2_Modelo1))
print(R)
```

```python
#Declaramos las variables dependientes e independientes para la regresi√≥n No lineal
Vars_Indep= df[['price']]
Var_Dep= df['host_total_listings_count']

#Redefinimos las variables 
x= Vars_Indep
y= Var_Dep

def  func1 (x, a, b, c):
    return a/b*x**2 + c*x #(Funci√≥n polinomial inversa)


#Ajustamos los par√°metros de la funci√≥n curve_fit
parametros1, covs = curve_fit(func1, df['price'], df['host_total_listings_count'])

# Predicciones
y_pred1 = func1(x, *parametros1)


#Graficamos las predicciones y los datos originales para realizar la comparaci√≥n
plt.plot(x, y, 'bo', label="y-original")
plt.plot(x, y_pred1, label="y-predecida con y=a/b*x**2 + c*x")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', fancybox=True, shadow=True)
plt.grid(True)
plt.show() 

#Calculamos el coeficiente de determinaci√≥n del modelo
R2_Modelo1 = r2_score(y, y_pred1) 
print(R2_Modelo1)

#Calculamos el coeficiente de correlaci√≥n del modelo
R = np.sqrt(abs(R2_Modelo1))
print(R)
```
